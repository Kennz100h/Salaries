##### 5. Find The Minimum,Maximum and Average BasePay

```python
# Calculate the minimum, maximum, and average BasePay
min_base_pay = df['BasePay'].min()
max_base_pay = df['BasePay'].max()
avg_base_pay = df['BasePay'].mean()

# Data for plotting
categories = ['Minimum BasePay', 'Average BasePay', 'Maximum BasePay']
values = [min_base_pay, avg_base_pay, max_base_pay]

# Create the bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(categories, values, color=['blue', 'green', 'red'])

# Add value labels on top of the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.5, round(yval, 2), ha='center', va='bottom')

# Add titles and labels
plt.title('Minimum, Average, and Maximum BasePay')
plt.xlabel('Category')
plt.ylabel('BasePay Amount')
plt.ylim(0, max_base_pay + 10000)  # Adjust the y-axis limit for better visualization

# Show theÂ plot
plt.show()
```

##### 10. Display Name Of The Person Having The Highest Basepay

```python
# Ensure 'BasePay' is numeric, coerce errors to NaN
df['BasePay'] = pd.to_numeric(df['BasePay'], errors='coerce')

# Drop rows where 'BasePay' is NaN
df = df.dropna(subset=['BasePay'])

# Sort the DataFrame by 'BasePay' in descending order and select the top 10 entries
top_10_basepay = df.nlargest(10, 'BasePay')

# Extract the employee names and their corresponding base pay
employee_names = top_10_basepay['EmployeeName']
base_pays = top_10_basepay['BasePay']

# Create the bar chart
plt.figure(figsize=(10, 6))
plt.barh(employee_names, base_pays, color='skyblue')
plt.xlabel('Base Pay ($)')
plt.ylabel('Employee Name')
plt.title('Top 10 Employees with Highest Base Pay')
plt.gca().invert_yaxis()  # Invert y-axis to have the highest base pay at the top
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()

# Display the bar chart
plt.show()
```

##### 11. Find The average Basepay of all Employees per Year

```python
# Ensure 'Year' and 'BasePay' are correctly formatted
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')  # Convert to numeric
df['BasePay'] = pd.to_numeric(df['BasePay'], errors='coerce')  # Convert to float

# Group by 'Year' and calculate the average BasePay
avg_basepay_per_year = df.groupby('Year')['BasePay'].mean().dropna()  # Remove NaN values

# Display the results
print(avg_basepay_per_year)
```

```python
# Plot a bar chart for Average BasePay per Year
plt.figure(figsize=(8, 5))
plt.bar(avg_basepay_per_year.index, avg_basepay_per_year.values, color='blue', alpha=0.7)

# Labeling
plt.xlabel("Year")
plt.ylabel("Average Base Pay ($)")
plt.title("Average BasePay of Employees Per Year")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the bar chart
plt.show()
```

```python
# Line chart for better trend visualization
plt.figure(figsize=(8, 5))
plt.plot(avg_basepay_per_year.index, avg_basepay_per_year.values, marker='o', linestyle='-', color='blue')

# Labeling
plt.xlabel("Year")
plt.ylabel("Average Base Pay ($)")
plt.title("Average BasePay of Employees Per Year")
plt.xticks(rotation=45)
```
